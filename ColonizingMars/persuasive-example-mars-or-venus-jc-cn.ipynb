This notebook compares the planets Mars and Venus in terms of physical characteristics as well as cultural popularity. We chose this challenge in order to make a recommendation about which planet humanity should consider colonizing.

## Physical Comparisons

Using [Martian](https://nssdc.gsfc.nasa.gov/planetary/factsheet/marsfact.html) and [Venutian](https://nssdc.gsfc.nasa.gov/planetary/factsheet/venusfact.html) data from NASA, we can compare these two planets to the Earth. These data sets were chosen so that we could make physical comparisons between the planets.

### Escape Velocity

```python
import matplotlib.pyplot as plt
%matplotlib inline
planets = ['Venus', 'Earth', 'Mars']
colors = ['yellow', 'blue', 'red']
escape_speeds = [10.36, 11.19, 5.03]
plt.bar(planets, escape_speeds, color=colors)
plt.title('Planetary Escape Velocities')
plt.ylabel('Escape Velocity (km/s)')
plt.show()
```

While Venus and Earth have similar gravity, Mars has a much lower escape velocity. The **advantage** of the lower escape velocity on **Mars** is the decreased cost of launching from the surface for return trips or the shipment of resources back to Earth.

### Day Lengths

```python
day_lengths = [2802, 24, 24.6597]
plt.bar(planets, day_lengths, color=colors)
plt.title('Planetary Day Lengths')
plt.ylabel('Day Length (h)')
plt.show()
```

Earth and Mars have similar day lengths, but the significantly longer day length on Venus may be difficult for humans to adapt to. This is another reason to consider Mars as a superior choice for colonization.

### Atmospheric Composition and Surface Pressure

```python
gasses = ['Carbon Dioxide', 'Nitrogen']
venus_atmosphere = [96.5, 3.5]
plt.pie(venus_atmosphere, labels=gasses)
plt.title('Gasses in Venutian Atmosphere')
plt.axis('equal')
plt.show()
```

```python
gasses = ['Carbon Dioxide', 'Nitrogen', 'Argon', 'Oxygen']
venus_atmosphere = [95.32, 2.7, 1.6, 0.13]
plt.pie(venus_atmosphere, labels=gasses)
plt.title('Gasses in Martian Atmosphere')
plt.axis('equal')
plt.show()
```

```python
surface_pressures = [92, 1.013, 0.00636]
plt.bar(planets, surface_pressures, color=colors)
plt.title('Planetary Surface Atmospheric Pressure')
plt.ylabel('Atmospheric Pressure (bar)')
plt.show()
```

Venus and Mars have similar compositions in terms of gasses present, mostly carbon dioxide. However the surface pressure on Venus is approximately 92 times that of Earth and is very inhospitable to life. As well, sulfuric acid rain on Venus would be problematic. Again, this makes Mars a more attractive choice.

## Cultural Comparisons

### Search Queries

Google Trends has data about keywords that people are searching for. We will look at [search frequency](https://trends.google.com/trends/explore?date=all&geo=CA&q=Mars,Venus) for the terms "Mars" and "Venus".

```python
!pip install pytrends --user            
from pytrends.request import TrendReq
import pandas as pd
import cufflinks as cf
cf.go_offline()

pytrend = TrendReq()
pytrend.build_payload(kw_list=['Mars', 'Venus'], timeframe='today 5-y')
df = pytrend.interest_over_time()
df.drop(columns=['isPartial']).iplot(title='Google Search Term Frequency', yTitle='Number of Searches', xTitle='Time')
```

We can see that "Mars" is usually a more common search term than "Venus", meaning that it is like more popular. There were a few days where Venus was a more common term, but let's compare the averages over this time period.

```python
print('Mars search frequency average:',df['Mars'].mean())
print('Venus search frequency average:',df['Venus'].mean())
```

It appears that, on average, Mars is searched for more than twice as often as Venus. However it may be that Google users are searching for those terms in a context other than the planets, such as the artist [Bruno Mars](https://en.wikipedia.org/wiki/Bruno_Mars). The Wikipedia disambiguation pages for [Mars](https://en.wikipedia.org/wiki/Mars_(disambiguation)) and [Venus](https://en.wikipedia.org/wiki/Venus_(disambiguation)) may provide some insight.

### Published Books

Using data from [Google Ngram Viewer](https://books.google.com/ngrams) we can compare mentions of the two planets in books.

```python
search_terms = ['Mars', 'Venus']
options = ['case_insensitive=on', 'year_start=1957', 'year_end=2008', 'corpus=15', 'smoothing=3']

content = '%2C'.join(search_terms)
options.insert(0, content)
parameters = '&'.join(options)
ngram_url_base = 'https://books.google.com/ngrams/interactive_chart?content='
ngram_url = ngram_url_base + parameters
print('Mentions of Mars or Venus in English Books Over Time')
#print(ngram_url)

from IPython.display import IFrame
IFrame(ngram_url, width=900, height=500)
```

Based on this data set from the beginning of the [space age](https://en.wikipedia.org/wiki/Space_Age), it appears that Mars has become more popular than Venus in published works.

# Conclusion

Based on physical and cultural comparisons, it is our recommendation that humanity should colonize Mars rather than Venus. Mars would appear to be more hospitable to human life in terms of day length and atmospheric conditions, and the lower surface gravity would make launches from Mars easier. As well, mentions of Mars seem to be more culturally prevalent leading to the assumption that Mars would be a more popular choice.

Of course more study and measurements of both planets is warranted before sending humans to either planet.
