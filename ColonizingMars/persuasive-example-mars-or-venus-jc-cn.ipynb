# Should We Colonize Mars or Venus?

### By John Carter and Carson Napier and Edgar Burroughs

This notebook compares the planets Mars and Venus in terms of physical characteristics as well as cultural popularity. The aim is to make a recommendation about which planet humanity should consider colonizing.

## Physical Comparisons

Using [Martian](https://nssdc.gsfc.nasa.gov/planetary/factsheet/marsfact.html) and [Venutian](https://nssdc.gsfc.nasa.gov/planetary/factsheet/venusfact.html) data from NASA, we can compare these two planets to the Earth.

### Escape Velocity

While Venus and Earth are similar in these respects, Mars has a much lower escape velocity. The advantage of the lower escape velocity on Mars is the decreased cost of launching from the surface for return trips or the shipment of resources back to Earth.

```python
import matplotlib.pyplot as plt
%matplotlib inline
planets = ['Venus', 'Earth', 'Mars']
colors = ['yellow', 'blue', 'red']
escape_speeds = [10.36, 11.19, 5.03]
plt.bar(planets, escape_speeds, color=colors)
plt.title('Planetary Escape Velocities')
plt.ylabel('Escape Velocity (km/s)')
plt.show()
```

### Day Lengths

Earth and Mars have similar day lengths, but the significantly longer day length on Venus may be difficult for humans to adapt to.

```python
day_lengths = [2802, 24, 24.6597]
plt.bar(planets, day_lengths, color=colors)
plt.title('Planetary Day Lengths')
plt.ylabel('Day Length (h)')
plt.show()
```

### Atmospheric Composition and Surface Pressure

Venus and Mars have similar compositions in terms of gasses present, mostly carbon dioxide. However the surface pressure on Venus is approximately 92 times that of Earth and is very inhospitable to life. As well, sulfuric acid rain on Venus would be problematic.

```python
gasses = ['Carbon Dioxide', 'Nitrogen']
venus_atmosphere = [96.5, 3.5]
plt.pie(venus_atmosphere, labels=gasses)
plt.title('Gasses in Venutian Atmosphere')
plt.axis('equal')
plt.show()
```

```python
gasses = ['Carbon Dioxide', 'Nitrogen', 'Argon', 'Oxygen']
venus_atmosphere = [95.32, 2.7, 1.6, 0.13]
plt.pie(venus_atmosphere, labels=gasses)
plt.title('Gasses in Martian Atmosphere')
plt.axis('equal')
plt.show()
```

```python
surface_pressures = [92, 1.013, 0.00636]
plt.bar(planets, surface_pressures, color=colors)
plt.title('Planetary Surface Atmospheric Pressure')
plt.ylabel('Atmospheric Pressure (bar)')
plt.show()
```

### Published Books

Using data from [Google Ngram Viewer](https://books.google.com/ngrams) we can compare mentions of the two planets in books. Based on this data set from the beginning of the [space age](https://en.wikipedia.org/wiki/Space_Age), it appears that Mars has become more popular than Venus in published works.

```python
search_terms = ['Mars', 'Venus']
options = ['case_insensitive=on', 'year_start=1957', 'year_end=2008', 'corpus=15', 'smoothing=3']

content = '%2C'.join(search_terms)
options.insert(0, content)
parameters = '&'.join(options)
ngram_url_base = 'https://books.google.com/ngrams/interactive_chart?content='
ngram_url = ngram_url_base + parameters
print('Mentions of Mars or Venus in English Books Over Time')
#print(ngram_url)

from IPython.display import IFrame
IFrame(ngram_url, width=900, height=500)
```
